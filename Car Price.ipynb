```python
plt.hist(df["Price"] , bins = 20  , edgecolor = "Blue" , color = "Orange")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x = df["Mileage"] , y = df["Price"])
sns.title("Price Vs Mielage")
plt.show()
```

```python
sns.lineplot(x = df["Year"] , y = df["Price"])
plt.title("Car Price Trend over the Year") 
plt.show()
```

```python
df.groupby("Brand")["Price"].mean().sort_values().plot(kind = "bar")
df.title("Avg car Price by Brand")
plt.ylabel("Avg Price")
plt.xticks(rotataion = 90)
plt.show()
```

```python
sns.boxplot(x=df['Transmission'], y=df['Price'])
plt.title('Transmission Type vs Price')
plt.show()
```

```python
sns.boxplot(x = df["Owner_Count"] , y = df["Price"])
plt.title("Owner count by Price")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = df , x = "Brand")
```

```python
top_brands_per_year = df.groupby(['Year', 'Brand']).size().reset_index(name='Sales')
print(top_brands_per_year.head())
```
